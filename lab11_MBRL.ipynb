# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(200):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=300) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

# MPC


```python
import numpy as np
import matplotlib.pyplot as plt

def mpc_random_shooting_step(gps, state, H=15, K=200, action_low=-1.0, action_high=1.0,
                             goal_pos=0.45, control_cost=1e-2, vel_cost=0.0, rng=None):
    """
    One MPC step using random shooting:
      - gps: list of GPManager objects [gp_for_dpos, gp_for_dvel]
      - state: np.array([pos, vel])
      - H: horizon
      - K: number of sampled action sequences
    Returns:
      best_action (float), best_action_sequence (H,), diagnostics dict
    """
    if rng is None:
        rng = np.random.RandomState()

    # Sample K sequences of H actions uniformly in [action_low, action_high]
    A = rng.uniform(action_low, action_high, size=(K, H)).astype(np.float32)

    # For each sequence, roll forward using GP mean dynamics
    # We'll maintain arrays shape (K, 2) for states
    states = np.tile(state.astype(np.float32), (K, 1))  # K x 2

    total_cost = np.zeros(K, dtype=np.float32)

    for t in range(H):
        # build input for GP: [pos, vel, action] for all K sequences at step t
        actions_t = A[:, t].reshape(-1, 1)
        X_t = np.hstack([states, actions_t])  # K x 3

        # Predict Δpos and Δvel using gps[0], gps[1]
        # gp.predict accepts arrays of shape (N, D)
        dpos_mean, dpos_var = gps[0].predict(X_t)
        dvel_mean, dvel_var = gps[1].predict(X_t)

        # Update states: s_next = s + Δs
        states[:, 0] += dpos_mean
        states[:, 1] += dvel_mean

        # Cost: encourage reaching goal_pos (final step more important), penalize control
        # Here: per-step cost = squared distance to goal in position + small control cost
        pos_error = (states[:, 0] - goal_pos)  # K
        step_cost = pos_error ** 2
        # optional velocity penalty
        if vel_cost > 0:
            step_cost += vel_cost * (states[:, 1] ** 2)
        # control cost for using large actions
        step_cost += control_cost * (actions_t[:, 0] ** 2)

        total_cost += step_cost

    # Also add a terminal bonus/penalty proportional to final distance (heavier)
    terminal_pos_error = (states[:, 0] - goal_pos) ** 2
    total_cost += 10.0 * terminal_pos_error

    best_idx = int(np.argmin(total_cost))
    best_seq = A[best_idx]
    best_action = float(best_seq[0])

    diagnostics = {
        'total_costs': total_cost,
        'best_index': best_idx,
        'best_sequence': best_seq,
        'predicted_final_states_of_best': states[best_idx],
    }
    return best_action, best_seq, diagnostics


def run_mpc_episode(env, gps, H=15, K=400, max_steps=1000,
                    action_low=-1.0, action_high=1.0,
                    render=False, rng_seed=0):
    """
    Run a single episode using MPC controller.
    Returns: trajectory dict with lists of states, actions, rewards, success flag
    """
    rng = np.random.RandomState(rng_seed)
    obs, _ = env.reset(seed=rng_seed)
    state = np.array(obs, dtype=np.float32)  # [pos, vel]
    traj = {'states': [], 'actions': [], 'rewards': []}
    done = False
    success = False
    goal_pos = 0.45

    # Warm-start: keep previous best sequence for shifting (optional)
    prev_best_seq = None

    for step in range(max_steps):
        # MPC planning
        best_action, best_seq, diag = mpc_random_shooting_step(
            gps, state, H=H, K=K, action_low=action_low, action_high=action_high,
            goal_pos=goal_pos, control_cost=1e-3, vel_cost=0.0, rng=rng
        )

        # Execute first action (env expects shape e.g. array([a]))
        action_to_env = np.array([best_action], dtype=np.float32)
        next_obs, reward, terminated, truncated, info = env.step(action_to_env)
        next_state = np.array(next_obs, dtype=np.float32)

        # Logging
        traj['states'].append(state.copy())
        traj['actions'].append(float(best_action))
        traj['rewards'].append(float(reward))

        state = next_state

        if render:
            env.render()

        # Check success (goal reached)
        if state[0] >= goal_pos:
            success = True
            print(f"Reached goal at step {step} pos={state[0]:.3f}")
            break

        if terminated or truncated:
            print(f"Environment done at step {step} (term={terminated}, trunc={truncated})")
            break

    traj['states'] = np.array(traj['states'])
    traj['actions'] = np.array(traj['actions'])
    traj['rewards'] = np.array(traj['rewards'])
    traj['success'] = success
    return traj


# -----------------------------
# Run MPC episode and plot
# -----------------------------
# Assumes `env` and `gps` exist and gps has been trained (from your code above)
traj = run_mpc_episode(env, gps, H=20, K=600, max_steps=500, rng_seed=42, render=False)

positions = traj['states'][:, 0]
velocities = traj['states'][:, 1]
actions = traj['actions']

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(positions, label='position')
plt.axhline(0.45, color='k', ls='--', label='goal pos=0.45')
plt.xlabel('timestep')
plt.ylabel('position')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(actions, label='action (applied)')
plt.xlabel('timestep')
plt.ylabel('action')
plt.legend()
plt.grid(True)

plt.suptitle(f"MPC episode — success={traj['success']}, steps={len(actions)}")
plt.tight_layout()
plt.show()
```

A short explanation (2–3 sentences) describing how MBRL is more efficient than Q-learning.

*   MBRL is more effeicent than Q-learning because it does not require constant actions in the real world. It is able to use a dynamics environment model to simulate and learn without taking real actions, which is much less costly than Q-learning, where is requires real experience. In this case, it was able to solve the mountain car problem in only 1 episode, which would not be possible with Q-learning, which is not model-based.


Describe the downside of MBRL over Q-learning

*   A downside to MBRL is that its accuracy is heavily dependant on its dynamics model, which can lead to inefficient or incorrect learning if the model is not accurate. With Q-learning, since it requires real actions, it does not need to depend on the accuracy of the model and learns directly in the real environment.
